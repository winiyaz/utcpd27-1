# Args and Kwargs Work 

## Function that add any number of arguments in function 
- below note that 

```python
# Create function any number of args and add it 

def addMany(*args):
    sum = 0
    for n in args:
        print(f'Entered numbers {n}')
        sum += n
    return print(f'Sum is : {sum}')

# Using list comprehension to create array of random numbers 
numbers = [i for i in range(1, 51)]
        
# Testing the function 
addMany(*numbers)
```

# `**kwargs` - Many keywords and Arguments

kwARGS = Keyword Arguments

```python
# kwargs - keywordarguments example
def calc1(**kwargs):
    print(kwargs)
    for k, v in kwargs.items():
        print(k)
        print(v)
    
    print(f'Kwargs are = {kwargs["add"]}')
    
calc1(add=3, multiply=4)
```

```python
# Calculate function again wiht kwarfs 

def calc2(n,**kwargs):
    print(kwargs)
    n += kwargs["add"]
    n *= kwargs['mul']
    print(n)
    
    
calc2(2, add=3, mul=5)
```
